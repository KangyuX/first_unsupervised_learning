```python
# K-Means Clustering

from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import silhouette_score

plt.figure()
X = np.array(df[["RESIDENTIAL Revenue Thousand Dollars", "COMMERCIAL Revenue Thousand Dollars"]])
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(X)
plt.scatter(df["RESIDENTIAL Revenue Thousand Dollars"], df["COMMERCIAL Revenue Thousand Dollars"], c=kmeans.labels_)
db = davies_bouldin_score(X, kmeans.labels_)
print("Davies-Bouldin Score:", db)
score = silhouette_score(X, kmeans.labels_)
print("Silhouette Score:", score)

plt.figure()
X = np.array(df[["RESIDENTIAL Revenue Thousand Dollars", "RESIDENTIAL Sales Megawatthours"]])
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(X)
plt.scatter(df["RESIDENTIAL Revenue Thousand Dollars"], df["RESIDENTIAL Sales Megawatthours"], c=kmeans.labels_)
db = davies_bouldin_score(X, kmeans.labels_)
print("Davies-Bouldin Score:", db)
score = silhouette_score(X, kmeans.labels_)
print("Silhouette Score:", score)
```

```python
# DBSCAN

from sklearn.cluster import DBSCAN

X = np.array(df[["RESIDENTIAL Revenue Thousand Dollars", "COMMERCIAL Revenue Thousand Dollars"]])
dbscan = DBSCAN(eps=5, min_samples=1)
dbscan.fit(X)
plt.figure()
plt.scatter(df["RESIDENTIAL Revenue Thousand Dollars"], df["COMMERCIAL Revenue Thousand Dollars"], c=dbscan.labels_)
```

```python
# Gaussian Mixture Model

import numpy as np
from sklearn.mixture import GaussianMixture
from matplotlib import pyplot as plt

X = np.array(df[["RESIDENTIAL Revenue Thousand Dollars", "COMMERCIAL Revenue Thousand Dollars"]])
gm = GaussianMixture(n_components=4, random_state=0).fit(X)

labels = gm.predict(X)
plt.figure()
plt.scatter(df["RESIDENTIAL Revenue Thousand Dollars"], df["COMMERCIAL Revenue Thousand Dollars"], c=labels)
plt.xlabel("RESIDENTIAL Revenue Thousand Dollars")
plt.ylabel("COMMERCIAL Revenue Thousand Dollars")
plt.title("Gaussian Mixture Clustering")
plt.show()
```
